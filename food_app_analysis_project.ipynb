```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

def read_zomato_dataset():
    hotels=pd.read_csv(r'C:\Users\Kissinger\Downloads\zomato.csv')
    return hotels
read_zomato_dataset()
```

```python
def remove_unwanted_columns():
    hotels = read_zomato_dataset()
    hotels.drop(['address','phone'], axis=1, inplace=True)
    return hotels
```

```python
def rename_columns():
    #rename columns,  only these columns are allowed in the dataset
    # 1.	Id
    # 2.	Name
    # 3.	online_order
    # 4.	book_table
    # 5.	rating
    # 6.	votes
    # 7.	location
    # 8.	rest_type
    # 9.	dish_liked
    # 10.	cuisines
    # 11.	approx_cost
    # 12.	type
    hotels = remove_unwanted_columns()
    hotels.columns = ['name','online_order','book_table','rating','votes','location','rest_type','dish_liked','cuisines','approx_cost','type']
    return hotels
```

```python
#handle  null values of each column
def null_value_check():
    hotels=rename_columns()
    
    #deleting null values of name column
    
    hotels.drop(hotels[hotels['name'].isna()].index, axis=0, inplace=True)
    
    #handling null values of online_order
    hotels['online_order'].fillna('NA', inplace=True)
    
    #handling null values of book_table
    hotels['book_table'].fillna('NA', inplace=True)
    
    #handling null values of rating
    hotels.drop(hotels[hotels['rating'].isna()].index, axis=0, inplace=True)
    
    #handling null values of votes
    hotels.drop(hotels[hotels['votes'].isna()].index, axis=0, inplace=True)
    
    #handling null values of location
    hotels['location'].fillna('NA', inplace=True)
    
    #handling null values of rest_type
    hotels['rest_type'].fillna('NA', inplace=True)
    
    #handling null values of dishliked
    hotels['dish_liked'].fillna('NA', inplace=True)
    
    #handling null values of cuisines
    hotels['cuisines'].fillna('NA', inplace=True)
    
    #handling null values of approxcost
    hotels.drop(hotels[hotels['approx_cost'].isna()].index, axis=0, inplace=True)
    
    #handling null values of type
    hotels['type'].fillna('NA', inplace=True)
    return hotels
null_value_check()
```

```python
#find duplicates in the dataset
def find_duplicates():   
    #call null_value_check() function to get dataframe
    hotels=null_value_check()
    #droping the duplicates value keeping the first
    hotels.drop_duplicates(keep='first',inplace=True)
  
    return hotels
```

```python
#removing irrelevant text from all the columns
def removing_irrelevant_text():
    #call find_duplicates() function to get dataframe
    hotels = find_duplicates()
    for column in hotels.columns:
        hotels.drop(hotels[hotels[column].str.contains('Rated', na=False)].index, axis=0, inplace=True)
        hotels.drop(hotels[hotels[column].str.contains('RATED', na=False)].index, axis=0, inplace=True)
    return hotels
removing_irrelevant_text()
```

```python
#check for unique values in each column and handle the irrelevant values
def check_for_unique_values():
    #call removing_irrelevant_text() function to get dataframe
    hotels=removing_irrelevant_text()
    
    hotels.drop(hotels[~(hotels['online_order'].str.contains('Yes') | hotels['online_order'].str.contains('No'))].index, axis=0, inplace=True)
    hotels['rating'] = hotels['rating'].str.replace('/5','')
    hotels.loc[hotels['rating']=='NEW', 'rating']=0
    hotels.loc[hotels['rating']=='-', 'rating']=0
    return hotels
```
