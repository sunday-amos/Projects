```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

def read_comments():
    comments=pd.read_csv(r'C:\Users\Kissinger\Downloads\comments.csv')
    return comments
read_comments()
```

```python
def clean_comments():
    #DO NOT REMOVE FOLLOWING LINE
    #call remove_unwanted_columns() function to get dataframe
    comments = read_comments()

    #Remove Unwanted columns
    comments.drop(['posted date', 'emoji used', 'Hashtags used count'], axis = 1, inplace = True)
    
    
    #rename columns, only these columns are allowed in the dataset
    # 1.	id
    # 2.	comment_text
    # 3.	user_id
    # 4.	photo_id
    # 5.	created_at
    
    comments.columns = ['id', 'comment_text', 'user_id', 'photo_id', 'created_at']
    
    
    #export cleaned Dataset to newcsv file named "comments_cleaned.csv"
    comments.to_csv(r'C:\Users\Kissinger\Downloads\comments_cleaned.csv', index = False)
    return comments
clean_comments()
```
